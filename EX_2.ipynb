```python
import hashlib


file_name = "evidence.txt"

def get_hash(file):
    with open(file, "rb") as f:
        data = f.read()
    return hashlib.sha256(data).hexdigest()

original_hash = get_hash(file_name)
print("Original Hash :", original_hash)

current_hash = get_hash(file_name)
print("Current Hash  :", current_hash)

if original_hash == current_hash:
    print("\nChain of Custody Verified!")
    print("The digital evidence has not been modified.")
else:
    print("\nWarning!")
    print("The digital evidence has been altered.")
```
