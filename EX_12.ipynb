```python
from PIL import Image

file = input("Enter image name: ")

try:
    image = Image.open(file)
    exif = image.getexif()

    if exif:
        print("EXIF Metadata:")
        for tag, value in exif.items():
            print(tag, ":", value)
    else:
        print("No EXIF metadata found.")

except FileNotFoundError:
    print("Image not found!")
```
