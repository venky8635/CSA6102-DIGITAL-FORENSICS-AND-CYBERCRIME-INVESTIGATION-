```python


file_name = input("Enter storage file name: ")

try:
    with open(file_name, "rb") as file:
        data = file.read()

    if b"\xFF\xD8\xFF" in data:
        print("JPEG file signature found.")
    if b"%PDF" in data:
        print("PDF file signature found.")
    if b"\x89PNG" in data:
        print("PNG file signature found.")

    if (b"\xFF\xD8\xFF" not in data and
        b"%PDF" not in data and
        b"\x89PNG" not in data):
        print("No recognizable file signatures found.")

except FileNotFoundError:
    print("File not found!")
```
