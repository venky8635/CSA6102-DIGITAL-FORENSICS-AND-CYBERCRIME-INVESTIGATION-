```python
import shutil
import hashlib

source = "evidence.txt"
copy = "evidence_copy.txt"

shutil.copyfile(source, copy)
print("Forensic copy created!")

def get_hash(file):
    with open(file, "rb") as f:
        return hashlib.sha256(f.read()).hexdigest()

source_hash = get_hash(source)
copy_hash = get_hash(copy)

print("Source Hash :", source_hash)
print("Copy Hash   :", copy_hash)

if source_hash == copy_hash:
    print("\nIntegrity Verified!")
    print("The forensic copy is identical to the original.")
else:
    print("\nIntegrity Check Failed!")
```
